# Dealing with large nums

```python
drinks['x'] = drinks.wine_servings * 1000
```

```python
drinks['y'] = drinks.total_litres_of_pure_alcohol * 1000
```

```python
drinks
```

```python
pd.set_option('display.float_format', '{:,}'.format)
# the python format str is passed and that is saying to 
# use commas as thousands separators
```

```python
drinks.head()
```

### But we need to deal with the x(int) coz the float_format only affects floating point columns 

```python
drinks['x'] = drinks.x.astype(float)
pd.set_option('display.float_format', '{:,}'.format)
drinks.head()
```
