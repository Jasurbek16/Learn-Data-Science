```python
people = {
    "first": ['Jack', 'James', 'Liam'],
    "last": ['Simons', 'Simons', 'Payne'],
    "email":['jsim@demo.com', 'jkab@demo.com', 'lpayne@demo.com']
}
```

```python
import pandas as pd

df = pd.DataFrame(people)
df
```

## Using filters

```python
filt = (df['last'] == 'Simons')
# ^ series obj

df[filt]
# ^ filtered one

###################### Alt:
df[df['last'] == 'Simons']

###################### Alt:
df.loc[filt, 'email']
```

## Using & and | operators for filtering multiple conditions

```python
filt_2 = (df['last'] == 'Simons') & (df['first'] == 'James')
df.loc[filt_2]
```

```python
filt_3 = (df['last'] == 'Payne') | (df['first'] == 'James')
df.loc[filt_3]
```

## Using ~ for negation

```python
filt_3 = (df['last'] == 'Payne') | (df['first'] == 'James')
df.loc[~filt_3]
# getting the opposite of above filter
```

## Filtering using a list of items

```python
high_salary = (df['ConvertedComp'] > 1000000)

countries = ['United States', 'United Kingdom', 'Uzbekistan', 'Russia']
filt_countries = (df['Country'].isin(countries))

filt_both = filt_countries & high_salary
```

# How to filter rows of a pandas Dframe by col value?

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
is_long = pd.Series(booleans)
# ^ converts booleans into pandas series
is_long.head()
```

```python
movies[is_long]
# ^ shows all of the True containing rows
```

# But the code above is not the efficient one

```python
is_long = movies.duration >= 200
# ^ this code replaces the for loop
is_long.head()
```

```python
# movies[is_long]
# Alt to the above:
movies[movies.duration >= 200]
```

```python
movies.loc[movies.duration >= 200, 'genre']
```
