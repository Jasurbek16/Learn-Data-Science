```python
people = {
    "first": ['Jack', 'James', 'Liam', 'Steve'],
    "last": ['Simons', 'Simons', 'Payne', 'Rogers'],
    "email":['jsim@demo.com', 'jkab@demo.com', 
             'lpayne@demo.com', 'captain@demo.com']
}

import pandas as pd

df = pd.DataFrame(people)
df
```

## Updating col.s

```python
df.columns
```

```python
df.columns = ['First_Name', 'Last_Name', 'Email']
```

```python
df
```

## Using a list comprehension for updating col.s

```python
df.columns = [x.lower() for x in df.columns]
df
```

## Using a str's replace method for replacing col patterns

```python
df.columns = df.columns.str.replace('_', '-')
df
```

## Changing not all of the col patterns

```python
df.rename(columns={'first-name':'first', 'last-name':'last'}, inplace=True)
```

```python
df
```

## Updating a row's col patterns

```python
df.loc[2] = ['Jack', 'Bond', 'jbon@demo.com']
# conditionals are also possible
df
```

```python
df.loc[2, ['last', 'email']] = ['Bobbis', 'jbobs@demo.com']
df
```

```python
df.loc[2, 'last'] = 'Bobb'
df
```

## using 'at' for looking up or changing a single value

```python
df.at[2, 'last'] = 'Sky'
df
```

## A common mistake

```python
# mistake
filt = (df['email'] == 'jkab@demo.com')
df[filt]['last'] = 'Smith'
```

```python
df.loc[filt, 'last'] = 'Smith'
df
```

## Applying a str's lower() for all of the rows of a col

```python
df.loc[0, 'email'] = 'jSiM@demO.com'
df['email'] = df['email'].str.lower()
df
```

<h3>apply</h3> 
<p>for calling a function on our values; can be used with both dataframes and series; runs a passed func to each value of a col when using with series; runs a func on each row or col of a dataframe when used with dataframes</p>


```python
# using apply with series. 
# That can apply a function to every value in the series
df['email'].apply(len)
```

## Using our own defined func

```python
def update_email(email):
    return email.upper()
df['email'] = df['email'].apply(update_email)
df
```

## Lambda Functions
<p>anonymus func.s without specific names </p>
<p>would work on a series obj</p>


```python
df['email'] = df['email'].apply(lambda x: x.lower())
df
```

## using the apply to the dframe
<p>runs for every col <br>
   can have 2nd arg as axis='columns' -> tells to count for each row the # of col.s
</p>

```python
df.apply(len)
```

```python
# ^ that used method is similar to:
len(df['email'])
# ^ but this would give us the result for all of the rows of one column
```

```python
df.apply(pd.Series.min)
# ^ gives the min val for each of the col.s
```

```python
df.apply(lambda x: x.min())
# x up there is a series not a value
```

## using an 'apply map' method for applying some func.s on each elem of the dataframe; 'apply map' only works with a dataframe not series

```python
df.applymap(len)
# ^ apply to every individual element of the dframe
```

```python
df.applymap(str.lower)
```

## the 'map' method only works on a series

```python
df['first'].map({'Jack': 'Sally', 'Steve':'Kalle'})
# NaN - not a number
```

## using a replace() for replacing only specified records
<p>NaN would not be in our way
</p>

```python
# if we are intending to replace some names not all, we use another method
df['first'] = df['first'].replace({'Jack': 'Sally', 'Jack':'Simon'})
df
```
