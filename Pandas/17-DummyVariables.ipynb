```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

# 1st way of dummy encoding 

```python
train['Sex_male'] = train.Sex.map({'female':0, 'male':1})
# ^ dummy encoding the col specified
```

```python
train.head()
```

# 2nd of dummy encoding

```python
pd.get_dummies(train.Sex)
# ^ creates a col for every possible value
```

## If we have "K" possible values for a categorical variable, we use "K - 1" dummy variables to represent that

### The dummy variable captures all possible info about the feature

```python
pd.get_dummies(train.Sex, prefix='Sex').iloc[:, 1:]
# the dropped col is defined as the "baseline"
# prefix -> adds a prefix with an _
```

```python
pd.concat([train, pd.get_dummies(train.Sex, prefix='Sex').iloc[:, 1:]], axis=1)
```

## Working with >2 possible values for a categorical variable

```python
train.Embarked.value_counts()
```

```python
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:, 1:]
```

```python
train = pd.concat([train, embarked_dummies], axis=1)
train.head()
```

## Passing in a dframe into the get_dummies()

```python
pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
# "columns" <- on which col.s I want the dummies
# the original col.s(passed into the columns) would 
# -be dropped and new ones would be added
# "drop_first=True" -> would care of dropping the additional 
# -col that we would have done manually otherwise
```
