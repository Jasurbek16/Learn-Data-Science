# SQL 
<p>We need to install a package for working with PostgreSQL:<br>
<strong>SQLAlchemy</strong> - (ORM-Object Relational Mapper) for working with dbs<br>
<strong>psycopg2-binary</strong></p>

## Writing

```python
from sqlalchemy import create_engine # for connecting to the db
import psycopg2
```

```python
engine = create_engine('postgresql://postgres:admin1605@localhost:5432/postgres')
```

```python
# writing to a table in the db
# the table gets created when we first refer to that
india_df.to_sql('sample_table', engine)
# cannot run more than 1 time coz the table is created already
```

## Updating

```python
india_df.to_sql('sample_table', engine, if_exists='replace')
# if_exists -> it's telling what it does
# ^ 1. giving an error is by def
# ^ 2. can append
# ^ 3. replacing
```

## Reading

```python
sql_df = pd.read_sql('sample_table', engine, index_col = 'Respondent')
# the name of the table is specified up there
```

```python
sql_df.head()
```

```python
sql_df = pd.read_sql_query('''SELECT * FROM sample_table 
                           WHERE "Age" > 15''', 
                           engine, 
                           index_col = 'Respondent')
# Instead of reading the entire table, we are going to run a query 
sql_df.head()
```
