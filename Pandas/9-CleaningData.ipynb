# Using numpy for specifying NaN values

```python
import pandas as pd
import numpy as np

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df = pd.DataFrame(people)

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
# ^ anytime we see those above, just replace

df
```

# Dropping entirely when seen any NaN for each row

```python
df.dropna()
```

# Default args of the .dropna()

```python
df.dropna(axis='index', how='any')
# ^ is using def args in back
# ^ axis='index' - drop rows containing missing values; 'columns' for col.s
# ^ how -> criteria used for dropping rows or col.s
# ^ how='any' drop rows containing any missing values
# ^ how='all' drop if all are missing
```

# Dropping specificly

```python
df.dropna(axis='index', how='any', subset = ['email'])
# subset -> a col that we are checking for missing values
```

# Dropping specificly according to multiple columns

```python
df.dropna(axis='index', how='all', subset = ['last', 'email'])
# ^ returns a result containing either last or email
# ^ 'all' -> both of the specified cols need to be missing for dropping
# ^ use 'inplace'
```

# Looking at a mask given by .isna()

```python
df.isna()
# ^ would give a mask of val.s that whether or not they are classified as NaN value
```

# Replace NaN values

```python
df.fillna('-')
# ^ replacing all of the missing val.s with the above str
```
