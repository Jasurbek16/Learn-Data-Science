## How do I use pandas with scikit-learn to create Kaggle submissions?

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
# the data that our model would be learning from
```

```python
train.head()
```

```python
# we need to create the feature matrix X -- (
# -features and col.s that the model learns from)
feature_cols = ['Pclass', 'Parch']
X = train.loc[:, feature_cols]
```

```python
X.shape 
```

```python
# responce/target vector creation(the thing we are going to predict)
y = train.Survived
y.shape
```

```python
from sklearn.linear_model import LogisticRegression # <- classification model
logreg = LogisticRegression()
# fitting the model to the training data
logreg.fit(X, y)
```

## Testing data to make predictions of

```python
test = pd.read_csv('http://bit.ly/kaggletest')
test.head()
```

```python
X_new = test.loc[:, feature_cols]
X_new.shape
```

```python
new_pred_class = logreg.predict(X_new)
# ^ predicted classes for the new data
```

```python
new_pred_class
```

```python
pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':new_pred_class}).set_index('PassengerId').to_csv('sub.csv')
```
