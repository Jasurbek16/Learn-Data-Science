```python
people = {
    "first": ['Jack', 'James', 'Liam'],
    "last": ['Simons', 'Kabby', 'Payne'],
    "email":['jsim@demo.com', 'jkab@demo.com', 'lpayne@demo.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
df['email']
```

```python
# setting email col as our unique index instead of num.s
df.set_index('email', inplace=True)
# inplace -> we need to tell pandas that we are changing
```

```python
df.index
# df.index.name = None <- could be used for clearing out the name of our custom index
```

```python
df.loc['jsim@demo.com', 'last']
# ^ used coz we changed the index
# '0' as above cannot be used coz that's not the index
```

```python
# resetting the changed index
df.reset_index(inplace=True)
df
# ^ if we have cleared out the name of our custom index col,
# -we have to rename that to it's original one and use the reset 
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.continent.head()
```

```python
drinks.set_index('country', inplace=True)
drinks.head()
```

```python
drinks.continent.head()
# index came from the dframe 
```

```python
drinks.continent.value_counts().values
```

```python
drinks.continent.value_counts()['Africa']
```

```python
drinks.continent.value_counts().sort_index()
# ^ sorting by the index
```

```python
people = pd.Series([3000000, 85000], index=['Albania', 'Andorra'], name='population')
people
```

```python
drinks.beer_servings * people
# ^ avg drink for a person in a year * the # of people
# ^ can be the example for the 'alignment'
```

```python
pd.concat([drinks, people], axis=1).head()
# axis -> tells to put obj.s side by side
# concat -> concatenates rows into rows or col.s into col.s
# because of the series index, the data has been put in the right spot
```
