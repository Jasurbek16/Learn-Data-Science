# Grouping 
<h3>Processes of Grouping</h3>
<ul>
    <li>Splitting an object
        <li>Applying func.s
            <li>Combining results
    </ul>

```python
country_grp = df.groupby(['Country'])
# the list in the groupby specifies a list of col.s to group by
# ^ contains groups of specfied col's values
```

# Applying funcs on each group

```python
country_grp.get_group('Uzbekistan').value_counts('OpSys', normalize=True)
# ^ getting the group of the UZB
```

```python
country_grp['OpSys'].value_counts(normalize=True).loc['South Korea']
```

```python
country_grp['ConvertedComp'].median().loc['Uzbekistan']
```

# Using agg() for applying multiple func.s

```python
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['South Korea']
```

# How much % of people know python from each country?

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
uses_py = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
uses_py
```

# Using concat() for combining series

```python
# we can combine multiple series together by using:
py_df = pd.concat([country_respondents, uses_py], axis='columns')
# Coz this would try to concat by rows, we specify by cols coz we 
# want by indexes
py_df
```

```python
py_df.rename(columns={'Country':'NumOfRespondents', 'LanguageWorkedWith':'NumKnowsPython'}, inplace=True)
```

```python
py_df
```

```python
py_df['%KnowsPython'] = (py_df['NumKnowsPython']/py_df['NumOfRespondents'] * 100)
py_df
```

```python
py_df.sort_values(by='%KnowsPython', ascending=False, inplace=True)
```

```python
py_df
```
