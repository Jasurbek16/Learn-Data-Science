## Using Pie Charts
### The percentages do not have to add up to 100

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(9, 6), dpi=80)
plt.style.use("fivethirtyeight")

slices = [60, 40, 30, 10]
labels = ['Slice1', 'Slice2', 'Slice3', 'Slice4']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})
# wedgeprops is being used for specifying the edge colors

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

## Using a Pie Chart for Plotting Popular Programming Languages

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(9, 6), dpi=80)
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
# ^ used for emphasizing some slice
# 0.1 -> 10% of the radius has been exploded out

plt.pie(slices, 
        labels=labels, 
        explode=explode, 
        shadow=True, 
        startangle=60,
        autopct='%1.1f%%',
        # ^ format string for using percentage
        wedgeprops={'edgecolor':'black'})
# wedgeprops is being used for specifying the edge colors
# ^ startangle is applied for the 1st item in the list for that's starting angle

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
